```python
#Setup
import pandas as pd
import numpy as np
```

```python
# Files to load
test_data_to_load = "Resources/test.csv"
```

```python
# Read the test data file and store it in a pandas DataFrame.
test_data_df = pd.read_csv(test_data_to_load)
test_data_df.head(10)
```

```python
zip_code_df = test_data_df.loc[:,"zip_code"]
zip_code_df.head(5)
```

```python
zip_code_df.to_csv('zip_code_test.csv', index=False)
```

```python
# Files to load
Coord_test = "Resources/zip_code_test_new.csv"
```

```python
# Read the test data file and store it in a pandas DataFrame.
coord_test_df = pd.read_csv(Coord_test)
coord_test_df.head(10)
```

```python
sort_new_df = test_data_df.sort_values(by=['zip_code'], ascending = True)
#len(sort_new_df)
sort_new_df.head()
```

```python
sorted_test_df = sort_new_df.reset_index(drop=True)
sorted_test_df.head()
```

```python
new_test_df = pd.concat([sorted_test_df, coord_test_df['Latitude'], coord_test_df['Longitude']],axis=1, join='outer')
new_test_df.head()
```

```python
new_df= new_test_df.loc[:,["beds","baths","size","lot_size", "lot_size_units", "price","zip_code", "Latitude", "Longitude"]]
new_df.head()
```

```python
new_df.to_csv('new_test.csv', index=False)
```
