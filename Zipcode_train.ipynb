```python
import pandas as pd
import os
```

```python
file = "Resources/train.csv"
train_df = pd.read_csv(file)
train_df
```

```python
# Clean the data
# Check for duplicates
train_df.isna().sum()
```

```python
#Drop the null values
train_df = train_df.dropna()
train_df.count()
```

```python
# Sort the dataframe in ascending order of zip codes
train_df = train_df.sort_values(by="zip_code", ascending=True, ignore_index=True)
train_df.head()
```

```python
# Creating a dataframe with just the zip code values
zip_df = pd.DataFrame(train_df["zip_code"])

zip_df.head()
```

```python
# Export the zip code dataframe to csv for further processing in excel
zip_df.to_csv("Zipcodes.csv", sep=',', index=None)
```

```python
file = "Zipcodes_updated.csv"
updated_zip = pd.read_csv(file)
updated_zip
```

```python
df_train = pd.concat([train_df, updated_zip["latitudes"], updated_zip["longitudes"]], axis=1, join="outer", ignore_index=False)
df_train.head()
```

```python
df_train = df_train[["beds","baths","size","size_units","lot_size","lot_size_units","zip_code","latitudes","longitudes","price"]]
df_train.head()
```
